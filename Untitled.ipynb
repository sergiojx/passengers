## Expliacion de la estrategia implentada
La estrategia expuesta, es el resultado de la experimentcion implementada el los bloques posteriores a este text, asi como al trbajo previo del profesor Daniel Jaramillo.
### Muestreo
El muestreo de la asceleracion de los tres ejes (x,y,n) se realiza cada 100mS. Y se opera sobre un buffer de 70 muestras.
### Estrategia de deteccion
Al inicial el sistema se recolectan 70 muestras iniciales, sobre las cuales se estiman la desviacion est치ndar y media. Cada 100mS se agrega una muestra nueva al buffer de cada eje, a la vez que se elimina la mas antiguia. Nuevamente se estime la desviacion est치ndar y la madia del nuevo conjunto. A partir de que se tienen 2 medidas (actual y previa) tanto de la desviacion standart y la media, se estima la tasa de variacion de cada una. En base a la magnitud y signo de la tasa de variacion de desviacion est치ndar se detecta un evento. En general un aumento (tasa de variacion positiva) repentino en la desviacion est치ndar inidicara una disprecion "repentina" de las muestras del buffer de cada uno de los ejes. Por otra parte el signo de la taza de variacion de la media, indicara el tipo de evento detectado.

static void __driverPro_SampleBuffIni(void)

static void __driverPro_Insert(ps_sampleBuffer_t buffer, int16_t val)
{
	if(buffer->ui8Index == 0)
	{
		buffer->int16Buff[0] = val;
		buffer->ui8Index = buffer->ui8Index + 1;
	}
	else
	{
		if(buffer->ui8Index == SAMPLE_WINDOW_ZISE)
		{
			// memset(shiftBuff,0,SAMPLE_WINDOW_ZISE);
			memcpy(shiftBuff, &buffer->int16Buff[1], (SAMPLE_WINDOW_ZISE - 1)*INT16_BYTE_SIZE);

			buffer->int16Buff[SAMPLE_WINDOW_ZISE - 1] = val;
		}
		else{
			buffer->int16Buff[buffer->ui8Index] = val;
			buffer->ui8Index = buffer->ui8Index + 1;
		}
	}

````python
def movingAverage(matrix, M):
    out = np.zeros(matrix.shape)
    lmbda = (M - 1)/M
    for i in range(matrix.shape[0] - 1):
        out[i + 1] = out[i]*lmbda + (1 - lmbda)*matrix[i]
        
    return out
````
